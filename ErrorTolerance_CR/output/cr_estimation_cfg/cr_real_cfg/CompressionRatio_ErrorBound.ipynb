# Compression Metrics Insights

```python
from matplotlib import rc
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.figure as fig
import numpy as np
import math

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['cm'],'size':30})
plt.rc('legend', fontsize=25)

def insight(name,NodeCount,HitRatio,TreeByteSize,EncodeSize,OutlierSize,CompressionRatio):
    
    f, axarr = plt.subplots(4,1,sharex=True,figsize = (20,15))
    plt.xticks(np.arange(12), (r'\textbf{1e-11}',r'\textbf{1e-10}',r'\textbf{1e-9}',r'\textbf{1e-8}',r'\textbf{1e-7}',
                               r'\textbf{1e-6}',r'\textbf{1e-5}',r'\textbf{1e-4}',r'\textbf{1e-3}',r'\textbf{1e-2}',
                               r'\textbf{1e-1}',r'\textbf{1e-0}'),rotation=30)

    ind = np.arange(12)
    
    axarr[0].plot(ind,CompressionRatio,'y*-',label=r"\textbf{CompressionRatio}",markersize=10,linewidth=3)
    axarr[0].set_yticks([0,50,100,150])
    axarr[0].set_yticklabels([r'\textbf{0}',r'\textbf{50}',r'\textbf{100}',r'\textbf{150}'])
    axarr[0].legend(loc='right', borderaxespad=0.)
    
    axarr[1].plot(ind,HitRatio,'bs-',label=r"\textbf{HitRatio}",markersize=10,linewidth=3)
    axarr[1].legend(loc='right', borderaxespad=0.)
    axarr[1].set_yticks([0,0.25,0.5,0.75,1])
    axarr[1].set_yticklabels([r'\textbf{0}',r'\textbf{0.25}',r'\textbf{0.50}',r'\textbf{0.75}',r'\textbf{1.0}'])
    
    axarr[2].plot(ind,NodeCount,'m8-',label=r"\textbf{NodeCount}",markersize=10,linewidth=3)
    axarr[2].legend(loc='right', borderaxespad=0.)
    axarr[2].set_yticks([0,10000,20000,30000,40000])
    axarr[2].set_yticklabels([r'\textbf{0}',r'\textbf{1e4}',r'\textbf{2e4}',r'\textbf{3e4}',r'\textbf{4e4}'])
    
    axarr[3].plot(ind,TreeByteSize,'r^-',label=r"\textbf{TreeByteSize}",markersize=10,linewidth=3)
    axarr[3].plot(ind,EncodeSize,'go-',label=r"\textbf{EncodeSize}",markersize=10,linewidth=3)
    axarr[3].plot(ind,OutlierSize,'bH-',label=r"\textbf{OutlierSize}",markersize=10,linewidth=3)
    axarr[3].plot(ind,TreeByteSize+EncodeSize+OutlierSize,'cs-',label=r"\textbf{TotalSize}",markersize=10,linewidth=3)
    axarr[3].set_yticks([0,100000,200000,300000,400000])
    axarr[3].set_yticklabels([r'\textbf{0}',r'\textbf{1e5}',r'\textbf{2e5}',r'\textbf{3e5}',r'\textbf{4e5}'])
    axarr[3].legend(loc='right', borderaxespad=0.)
    
    f.subplots_adjust(top=0.90)
    f.text(0.5, 0.05, r"\textbf{Error Bound}", ha='center')
    plt.savefig('crmetrics_errorbound_'+name+'.pdf',bbox_inches='tight')
#     plt.show()
    plt.clf()
```
